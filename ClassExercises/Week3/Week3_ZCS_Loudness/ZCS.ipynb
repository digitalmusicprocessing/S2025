```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
## Setup linear chirp (groups)
sr = 44100
t = np.arange(sr*4)/sr
ft = 220*t + 50*t**2
y = np.cos(2*np.pi*ft)
ipd.Audio(y, rate=sr)


## Look at zero crossings in different regions (me)
plt.plot(y[0:1000])
plt.plot(y[sr*2:sr*2+1000])
plt.plot(np.zeros(1000), c='k', linestyle='--')


## Compute zero crossings centered at each sample (groups)

## Do this more efficiently using cumulative sums (groups)

```
