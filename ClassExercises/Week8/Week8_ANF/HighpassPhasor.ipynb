Examine what a highpass filter

$y[n] = x[n] - x[n-1]$

does to a phasor 

$x[n] = \cos(2 \pi f n + \phi)$

instead work with

$x[n] = e^{i 2 \pi f n}$


$y[n] = e^{i 2 \pi f n} - e^{i 2 \pi f (n-1)}$

$y[n] = e^{i 2 \pi fn}(1 - e^{-i 2 \pi f })$

### $y[n] = e^{i 2 \pi fn}(\frac{e^{i \pi f} - e^{-i \pi f}}{e^{i \pi f}})$

### $y[n] = e^{i 2 \pi fn}(\frac{2 i \sin( \pi f)}{e^{i \pi f}})$

$A = 2 \sin( \pi f)$

```python
import numpy as np
import matplotlib.pyplot as plt
f = np.linspace(0, 0.5, 10000)
amps = (2*np.sin(np.pi*f))
plt.plot(f, amps)
```

```python
# DAW: Digital Audio Workspace 
# (Professional software to process digital audio)

def highpass(x):
    N = len(x)
    y = np.zeros(N)
    for n in range(1, N):
        y[n] = x[n] - x[n-1]
    return y

N = 1000
f = 0.03
x = np.cos(2*np.pi*f*np.arange(N))
y = highpass(x)
plt.subplot(211)
plt.plot(x)
plt.ylim([-1.1, 1.1])
plt.title("x, f = {}".format(f))
plt.subplot(212)
plt.plot(y)
plt.ylim([-2.2, 2.2])
plt.title("y")
plt.show()
```
